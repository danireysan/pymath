```python
import sympy as sym
from IPython.display import display, Math
```

```python
q = sym.symbols('q')

print(sym.diff( sym.cos(q)))
print(sym.diff( sym.sin(q)))
```

```python
f = sym.cos(q)

for i in range(0, 8):
    display(Math('\\frac{d}{dx}%s = %s' %(sym.latex(f), sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
import sympy.plotting.plot as symplot
x = sym.symbols('x')

f = sym.cos(q)

for i in range(0,4):
    if i == 0:
        p = symplot(f, show=False, label= sym.latex(f))
    else:
        p.extend(symplot(f, show=False, label= sym.latex(f)))
    f = sym.diff(f)

p.legend = True
p.xlim = [-3,3]
p.show()
```

```python
x, a = sym.symbols('x,a' )

fax = sym.cos(x + sym.sin(x)) + a

for ai in range(0,4):
    
    if ai == 0: 
        
        p = symplot(fax.subs(a, ai), show=False, label= "a=" + str(ai))
    else:
        p.extend(symplot(fax.subs(a, ai), show=False, label= "a=" + str(ai)))
p.title = 'The functions'
p.legend = True
p.show()

for ai in range(0,4):
    
    if ai == 0: 
        
        p = symplot(sym.diff(fax.subs(a, ai)), show=False, label= "a=" + str(ai))
    else:
        p.extend(symplot(sym.diff(fax.subs(a, ai)), show=False, label= "a=" + str(ai)))
        
p.title = 'Their derivatives'
p.legend = True
p.show()
```
