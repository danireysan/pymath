```python
import sympy as sym
from IPython.display import display, Math
```

```python
q = sym.symbols('q')

print(sym.diff( sym.cos(q)))
print(sym.diff( sym.sin(q)))
```

```python
f = sym.cos(q)

for i in range(0, 8):
    display(Math('\\frac{d}{dx}%s = %s' %(sym.latex(f), sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
import sympy.plotting.plot as symplot

f = sym.cos(q)

for i in range(0,4):
    if i == 0:
        p = symplot(f, show=False, label= sym.latex(f))
    else:
        p.extend(symplot(f, show=False, label= sym.latex(f)))
    f = sym.diff(f)

p.legend = True
p.xlim = [-3,3]
p.show()
```
