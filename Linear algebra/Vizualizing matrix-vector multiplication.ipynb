```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.array([ [.5,1], [1,.5]] )

v = np.array([1,2])
v = np.random.rand(2)

Mv = M@v

plt.plot([0,v[0]],[0,v[1]], 'b', linewidth=2 , label='v' )
plt.plot([0,Mv[0]],[0,Mv[1]], 'r--', linewidth=2 , label='Mv' )
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()


# The product of a matrix and vector is the vector changing in it's size and direction
```

```python
M = np.array([ [.5,1], [1,.5]] )

vcomp = np.linspace(-2,2,40)

for a in vcomp:
    v = np.array([1,a])
    Mv = M@v
    
    plt.plot([0,v[0]], [0,v[1]], color=[1-abs(a)/4,.5,abs(a)/2], alpha=.5)
    plt.plot([0,Mv[0]], [0,Mv[1]], color=[.5, 1-abs(a)/4, abs(a)/2 ], alpha=.5)
    
plt.axis('square')
plt.axis('off')
plt.show()
```
