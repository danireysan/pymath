```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ein = Image.open('einstein.jpg')
print(np.shape(ein))

ein = np.mean(ein, 2)

plt.imshow(ein)
plt.show()
```

```python
U,s, v = np.linalg.svd(ein)

fig, ax = plt.subplots(1,3)

ax[0].imshow(U)
ax[0].set_title("U")

ax[1].imshow(np.diag(np.log(s)))
ax[1].set_title("$\\Sigma$")

ax[2].imshow(v)
ax[2].set_title("s")

plt.show()
```

```python
plt.plot(s, 'ks-', markerfacecolor= 'w')
plt.xlim([-1,50])
plt.title("Eigenspectrum of Einstein")
plt.show()
```

```python
# Exercise 1: Reconstruct the image using single value decomposition multiplication
S = np.zeros(np.shape(ein))
for i in range(0, len(s)):
    S[i,i] = s[i]
    
rein = U@S@v

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('original matrix')

plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD image')

plt.show()
```

```python
# Exercise 2: Shuffle the order of singular values

randorder = np.random.permutation(len(s))


S = np.zeros(np.shape(ein))
for i in range(0, len(s)):
    S[i,i] = s[randorder[i]]
    
rein = U@S@v

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('original matrix')

plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD image')

plt.show()
```

```python
# Exercise 3: Reconstruct the image with only 10 components

randorder = np.random.permutation(len(s))


S = np.zeros(np.shape(ein))
for i in range(0, 20):
    S[i,i] = s[i]
    
rein = U@S@v

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('original matrix')

plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD image')

plt.show()
```
