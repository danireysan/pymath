```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1  = np.random.rand(50)
v2 = np.random.rand(80)

op = np.outer(v1,v2)

plt.imshow(op)
plt.show()
```

```python
v = np.arange(1,11)
w = np.arange(1,6)

print(np.outer(v,w))
print(np.outer(w,v))
```

```python
s = 4


res1 = s*np.outer(v,w)
res2 = np.outer(s*v,w)
res3 = np.outer(v,s*w)
res4 = np.outer(s,v) * s

print(res1-res2)
```
