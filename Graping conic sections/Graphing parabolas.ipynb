```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters
a = 1 # 1/4y
p = 1/(4*a)
h = 1 
k = -2
n = 100

# x-axis points to evaluate the function
x = np.linspace(h-2,h+2, n)

# create the function
y = a*(x-h)**2 + k

# plot the parabola
plt.plot(x,y)

# plot the vertex
plt.plot(h,k, 'ro', label='Vertex')

# plot the focus
plt.plot(h,k+p, 'go', label='Focus')

# plot the directrix
d = k-p
plt.plot(x[[0,-1]],[d,d], label='Directrix')

plt.grid()
plt.legend()
plt.show()
```
