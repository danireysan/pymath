```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1,2,2,2,2,4,5,6,1,2,9,4,5,8,9,2]

plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Number of apparences (count)')
plt.show()

# if a zero appears in the first array of this function it means that in the index that the zero appears and the next one
# there is no data between those two numbers
# histdata = np.histogram(data)
# Output of this function:
# (array([2, 6, 0, 2, 0, 2, 1, 0, 1, 2], dtype=int64), array([1. , 1.8, 2.6, 3.4, 4.2, 5. , 5.8, 6.6, 7.4, 8.2, 9. ]))
y,x = np.histogram(data, bins =np.arange(0,11))
yP = y/np.sum(y)

plt.bar(x[:-1], yP)
plt.xlabel('Value')
plt.ylabel('Number of apparences (count)')
plt.xticks(np.arange(0,10))
plt.show()
```

```python

y,x = np.histogram(data, bins =np.arange(0,11))
yP = y/np.max(y)

plt.bar(x[:-1], yP)
plt.xlabel('Value')
plt.ylabel('Number of apparences (norm.)')
plt.xticks(np.arange(0,10))
plt.show()
```
