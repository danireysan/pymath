```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [1,2,2,2,2,4,5,6,1,2,9,4,5,8,9,2]

plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Number of apparences (count)')
plt.show()

# if a zero appears in the first array of this function it means that in the index that the zero appears and the next one
# there is no data between those two numbers
# histdata = np.histogram(data)
# Output of this function:
# (array([2, 6, 0, 2, 0, 2, 1, 0, 1, 2], dtype=int64), array([1. , 1.8, 2.6, 3.4, 4.2, 5. , 5.8, 6.6, 7.4, 8.2, 9. ]))
y,x = np.histogram(data, bins =np.arange(0,11))
yP = y/np.sum(y)

plt.bar(x[:-1], yP)
plt.xlabel('Value')
plt.ylabel('Number of apparences (count)')
plt.xticks(np.arange(0,10))
plt.show()
```

```python
# Normalized
y,x = np.histogram(data, bins =np.arange(0,11))
yP = y/np.max(y)

plt.bar(x[:-1], yP)
plt.xlabel('Value')
plt.ylabel('Number of apparences (norm.)')
plt.xticks(np.arange(0,10))
plt.show()
```

```python
# Make a graph and probability density of this functions
# y = x, y = x^2, y = x^3
x = np.linspace(-3,3, 1000)

fig, ax = plt.subplots(2,3)
fig.set_size_inches(10,6)

for i in range(0,3):
    # create the functino
    y = x**(i+1)
    h = np.histogram(y,50)
    yP = h[0] / np.sum(h[0])
    
    # plotting
    ax[0,i].plot(x,y)
    ax[0,i].set_title('$y=x^{%g}$' %(i+1))
    # histogram
    
    ax[1,i].plot(h[1][:-1], yP)
```
