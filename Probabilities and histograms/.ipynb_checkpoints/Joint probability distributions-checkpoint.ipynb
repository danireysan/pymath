```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
n = 1000
    
x = np.random.randn(n)
y = np.random.randn(n)**2 

plt.hist2d(x,y,30, vmax=10)
plt.xlabel('x')
plt.ylabel('$x^2$')
plt.show()
```

```python
sns.set()

sns.jointplot(x=x,y=y)
plt.show()
```

```python
sns.set()

sns.jointplot(x=x,y=y, kind='reg')
plt.show()
```

```python
sns.set()

sns.jointplot(x=x,y=y, kind='kde')
plt.show()
```

```python
sns.set()

sns.jointplot(x=x,y=y, kind='scatter', color=[.4,.7,.2]).plot_joint(sns.kdeplot)
plt.show()
```

```python
# exercise
csv = pd.read_csv('data.csv',header=None, names=['x','y'])

sns.jointplot(x=csv['x'], y=csv['y'], kind='kde', levels= 10 ).plot_joint(sns.scatterplot)
plt.show()
```
