```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shift = 5 
stretch =2 
n =1000

pnts = np.random.randn(n)*stretch+ shift

print("Mean is %g, std is %g" %(np.mean(pnts), np.std(pnts)))

fig, ax = plt.subplots(1,2, figsize=(6,3))

ax[0].plot(pnts, 's', alpha=.4)
ax[1].hist(pnts,50)

plt.show()
```
