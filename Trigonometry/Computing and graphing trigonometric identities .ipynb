```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,6*np.pi, 90)

plt.plot(x, np.sin(x), 'r', label='$\\sin(\\theta)$')
plt.plot(x, np.cos(x), 'b', label='$\\cos(\\theta)$')

plt.legend()
plt.xlabel('rad.')
plt.ylabel('function value')
plt.show()
```

```python
th = np.linspace(0,4*np.pi, 30)

plt.plot(th, np.tan(th), 'k')
plt.ylim([-5,5])
plt.show()
```

```python
ang = np.random.rand()*2*np.pi

tan = np.tan(ang)
sc  = np.sin(ang) / np.cos(ang)

print(ang, tan - sc)
```

```python
# c**2 + x**2 = 1 
thetas = np.linspace(0, 2*np.pi, 13)
print(np.cos(thetas) * np.sin(thetas) ** 2) 
```
