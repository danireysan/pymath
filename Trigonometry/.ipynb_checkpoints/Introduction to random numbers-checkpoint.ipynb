```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand(1000)

minval = 2
maxval = 17

nums = nums*(maxval-minval) + minval

plt.plot(nums, 's', alpha=.5)
plt.show()


plt.hist(nums)
plt.show()
```

```python
nums = np.random.randn(1000)
plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```

```python

desired_mean = 15
desired_std  = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums)
nums = nums/np.std(nums)

nums = nums*desired_std + desired_mean



plt.subplot(121)
plt.plot(nums, 's')

plt.subplot(122)
plt.hist(nums)

plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' %(np.mean(nums), np.std(nums) ))
```

```python
angs = np.random.rand(100)*2*np.pi

for i in angs: 
    plt.plot([0, np.cos(i)], [0,np.sin(i)])
    
plt.axis('square')
plt.axis([-1,1,-1,1])
plt.axis('off')
plt.show()
```
