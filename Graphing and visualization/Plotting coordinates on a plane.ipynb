```python
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = 3
y = 5

plt.plot(x,y, 'kp')

plt.axis('square')
plt.axis([-6,6, -6, 6])
plt.grid()

plt.show()
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label='point %s' %(str(i+1)))
    
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4,2, 'r+')

axis = plt.gca()
ylim = axis.get_ylim()
axis.set_ylim([ylim[0], 3])
```

```python
x = range(-10, 11)

for xi in x:
    y = xi ** 2 - 3*xi
    plt.plot(xi, y, 'o')
axis = plt.gca()
xlim = axis.get_xlim()
axis.set_xlim([xlim[0], 11])
plt.grid()
plt.show()
```
