```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = mx + b 

x = [-5,5]
m = -.3
b = .1

#y = [340, 2235]
#for i in range(0, len(x)):
#    y[i] = m * x[i] + b


y = m*np.array(x) +b
plt.plot(x,y, label= 'y =%sx + %s ' %(m,b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
# y = mx + b 

x = [-5,5]
m = 0.7
b = -2
m2 = -1.25
b2 = 0.75
y = m*np.array(x) +b
y2 = m2*np.array(x) +b2
plt.plot(x,y, label= 'y =%sx + %s ' %(m,b))
plt.plot(x,y2, label= 'y =%sx + %s ' %(m2,b2))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
x = [-5, 5]
m = [.7, -1.25]
b = [-2, .75]

for i in range(0, len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y, label='y = %s + %s' %(m[i], b[i]))
    
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()
plt.show()
```
