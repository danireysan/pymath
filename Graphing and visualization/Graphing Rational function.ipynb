```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0, len(x) ):
    y[i] = 2 - x[i]**2

plt.plot(x,y, 's-')
plt.show()
    
```

```python
x = np.linspace(-3,4,100)
y1 = 2 + abs(x) ** (1/3)
y2 = 2 + abs(x) ** (1/2)


plt.plot(x,y1, 'm-')
plt.plot(x,y2, 'b-')
plt.show()
```

```python
x = np.linspace(-4,4,200)

for i in range(-1,4):
    y = x**i
    plt.plot(x, y, label="$y=x^{%s}$" %i, linewidth=4)
    
plt.legend()
plt.xlim([x[0], x[-1]])
plt.ylim([-20, 20])
plt.show()
```
