```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[1,2], [1,4]]
plt.imshow(A)

plt.xticks([0,1])
plt.yticks([0,1])

plt.show()
```

```python
A = np.zeros((10,14))

print( np.shape(A) )

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        
        # populate the matrix
        A[i,j] = 3*i-4*j

print(A)
plt.imshow(A)
plt.plot([0,3],[8,2],'r',linewidth=4)


for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')

plt.set_cmap('twilight_shifted')


plt.savefig('AnImageYes.pdf')

plt.show()
```

```python
# "soft-coding"
N = 10

C = np.zeros((N,N))

for i in range(0,N):
    for j in range(0,N):
        C[i,j] = (-1)**(i+j) #PEMDAS
        
plt.imshow(C)
plt.set_cmap('gray')
plt.xticks([])
plt.yticks([])
plt.show()
print(C)
```
