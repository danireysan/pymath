```python
import matplotlib.pyplot as plt
import math
```

```python

p1 = [-3, -1]
p2 = [4,4]

plt.plot([p1[0],p2[0]], [p1[1],p2[1]], color=[.4,.2,.7], linewidth=5)

plt.axis('square')
plt.axis([-6,6, -6,6])
plt.grid()
plt.show()
```

```python
x = 3
y = 5

plt.plot(x,y, 'ro')
plt.plot([0,x], [0,y], 'r')
plt.axis('square')
plt.axis([-6,6, -6,6])
plt.grid()
plt.plot([-6,6], [0,0], 'k')
plt.plot([0,0], [-6,6], 'k')

plt.show()
```

```python

x = range(-20, 20)

for xi in x:
    y = math.sqrt(abs(xi))
    plt.plot([0,xi], [0,y])
axis = plt.gca()
xlim = axis.get_xlim()
plt.grid()
plt.show()
```
