```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
n = 5

for k in range(0, n):
    z = sym.exp( 2*sym.pi*sym.I*k/n)
    display(Math('{%s}^{%s} \\Rightarrow %s' %(sym.latex(z),n, sym.latex(z**n))))
```

```python
n = 8

for k in range(0, n):
    z = np.exp( 2*np.pi*1j*k/n)
    plt.plot([0,np.real(z)],[0, np.imag(z)])
    
plt.axis('square')

x = np.linspace(0,2*np.pi, 100)
plt.plot(np.cos(x), np.sin(x), color='gray')

plt.show()
```

```python
n = 200

color = np.linspace(0,.7, n)

for k in range(0, n):
    z = k * np.exp( 2*np.pi*1j*k/n)
    plt.plot([0,np.real(z)],[0, np.imag(z)], linewidth=2, color=[color[k], color[k], 0.3])

plt.axis('square')
plt.axis('off')
plt.show()
```
