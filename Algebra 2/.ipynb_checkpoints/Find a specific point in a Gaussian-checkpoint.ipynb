```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = .999 

t = np.linspace(-3, 3, 1001)

# gaussian 
g = np.exp( -4*np.log(2)*t**2 / h**2 )

# plot!

plt.plot(t,g)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Gaussian with FWHM' + str(h) )
plt.show()
```

```python
yVal2find = .5

tvals = h*np.sqrt( np.log(yVal2find)/(-4*np.log(2)) )

print(tvals, -tvals)
```

```python
N = 100

G = np.zeros((N,len(t)))
h = np.zeros(N)

for i in range(N):
    h[i] = (i+1)/N
    G[i, :] = np.exp( -4*np.log(2)*t**2 / h[i]**2)
    
plt.pcolormesh(t,h,G, shading='auto')
plt.xlabel('t')
plt.ylabel('h')
plt.show()
```
