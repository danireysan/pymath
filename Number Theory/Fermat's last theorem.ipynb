```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ints = np.arange(1,101)

Z = np.zeros((len(ints), len(ints)))

for x in ints:
    for y in ints:
        Z[x-1,y-1] = (x**3 + y**3)**(1/3)
```

```python
zInts = Z%1 == 0
```

```python
fig, ax = plt.subplots(1,2, figsize= (12,6))

ax[0].imshow(Z)
ax[1].imshow(zInts)

plt.show()
```

# Fermat's last theorem

This theorem simply states that no three positive integers, a, b and c can satisfy the equation a^n+ b^n = c^n
in other words, there are no whole number solutions to the equation that is greater than 2
