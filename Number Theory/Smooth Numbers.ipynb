```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
maxN = 10000

largestPrimeFact = np.zeros(maxN+1, dtype=int)/0

for i in range(2, maxN+1):
    largestPrimeFact[i] = np.max(sym.primefactors(i))
    
plt.plot(largestPrimeFact, 'k.', markersize=.5)
plt.xlabel('Number')
plt.ylabel('Largest Prime Factor')
```

```python
# count the number of occurences of max prime factors

uniqueMaxPrimes = np.unique(largestPrimeFact)

counts = np.zeros((len(uniqueMaxPrimes)))
for i,u in enumerate(uniqueMaxPrimes):
    counts[i] = np.sum(largestPrimeFact ==u)
    
plt.plot(uniqueMaxPrimes,counts, ' k.')
plt.xlim([-2, 500])
plt.title('%s modal is the max prime factor' %uniqueMaxPrimes[np.argmax(counts)])
plt.show()
```

```python
numberList = np.arange(maxN+1)


numberList[largestPrimeFact<=5]
```

```python
smoothCount = np.zeros(len(uniqueMaxPrimes))
for i,u in enumerate(uniqueMaxPrimes):
    smoothCount[i] = len(numberList[largestPrimeFact<=u])
```

```python
plt.plot(uniqueMaxPrimes, smoothCount, 'k.')
plt.xlim([-2,500])
plt.xlabel('Smooth number')
plt.ylabel('Number of numbers')
plt.title('K-smooth')
plt.show()
```
