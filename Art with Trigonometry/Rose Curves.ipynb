```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 2.2

t = np.linspace(0,8 * np.pi, 1000 )

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,'m', linewidth=3)
plt.axis('square')
plt.axis('off')

plt.show()
```

```python
def clawso (number):
    k = number

    t = np.linspace(0,8 * np.pi, 1000 )

    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)

    plt.plot(x,y,'m', linewidth=3)
    plt.axis('square')
    plt.axis('off')

    plt.show()
    
for n in np.arange(0, 8, 0.5):
    clawso(n)
```
