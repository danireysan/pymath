```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-6*np.pi, 6*np.pi,1000)

a = t

x = a * np.cos(t)**3
y = a * np.sin(t)**3

plt.plot(x,y, 'm', linewidth=3)

plt.axis("off")

plt.show()
```

```python

# Code that I'm amazed it didn't crash
fig, ax = plt.subplots(3,3) 

row, col = np.indices((3,3))

def astroid(n):

    a = t**n

    x = a * np.cos(t)**3
    y = a * np.sin(t)**3
    
    r = row.ravel()[n]
    c = col.ravel()[n]
    
    return [r,c]
    
for n in range(9):
    my_list = astroid(n)
    r = my_list[0]
    c = my_list[1]
    ax[r,c].plot(x,y, 'b', linewidth=3)
    ax[r,c].axis("off")
    ax[r,c].axis("square")
    plt.show()
    
```
