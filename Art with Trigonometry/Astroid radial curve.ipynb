```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-6*np.pi, 6*np.pi,1000)

a = t

x = a * np.cos(t)**3
y = a * np.sin(t)**3

plt.plot(x,y, 'm', linewidth=3)

plt.axis("off")

plt.show()
```

```python
evenArray = [0,2,4,6,8]

def astroid(n):
    t = np.linspace(-6*np.pi, 6*np.pi,1000)

    a = t*n

    x = a * np.cos(t)**3
    y = a * np.sin(t)**3

    plt.plot(x,y, 'm', linewidth=3)

    plt.axis("off")
    plt.show()
    
for n in evenArray:
    astroid(n)
    
```
